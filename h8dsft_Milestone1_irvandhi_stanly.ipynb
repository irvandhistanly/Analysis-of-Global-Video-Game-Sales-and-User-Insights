Then, we want to observe is there a meaningful difference between sales in different region. This is important as we need to know which region we should focus our action game on, in terms of marketing, game theme, campaigns, etc. To determine whether their difference is significant or not, we should conduct multiple comparisons test. However, as it was found that the distribution of global sales for the action genre is not in normal distribution. Thus, we cannot conduct parametric tests, such as ANOVA and use the non-parametric alternatives, such as Kruskal-Wallis test. Both tests aim for the same thing, which is to determine if the mean difference of the sales in different regions is significant or not. To gain a general overview, the mean of each column was also calculated.

Below are the hypothesis of this test:

1. H0: The mean difference of sales in 4 different regions are not significant
2. H1: The mean difference of sales in 4 different regions are significant

From the Kruskal-Wallis Test, we can conclude that there is a significant difference between 4 of them as the p-value of the Kruskal-Wallis test is lower than 0.05, thus we fail to reject null hypothesis and there is a significant difference. Furthermore, we still want to observe pairwise comparison in between 4 regions (NA vs EU, JP vs Other, etc) to fully gain insight whether the the region with the highest sale is significantly different with the region with the 2nd highest sale. Thus, Dunn's Test with Bonferroni Correction was also conducted and it was found that the pairwise comparison also showed significant different among individual regions.

From this information, we can conclude that the gaming industry, especially the action-based one, will mostlikely generate higher profits in NA region, followed by Europe region. Thus, we will focus our marketing, game designs, cultural aspects of the game, campaigns, etc. accordingly to attract users from the North America to generate sales effectively.

```python
#Creating 4 variables that consist of the sales of video games within their respective regions
most_genre_NA = action_data['NA_Sales']
most_genre_EU = action_data['EU_Sales']
most_genre_JP = action_data['JP_Sales']
most_genre_Other = action_data['Other_Sales']

#Calculating the mean just for observation
print(most_genre_NA.mean())
print(most_genre_EU.mean())
print(most_genre_JP.mean())
print(most_genre_Other.mean())
```

```python
#Conducting Kruskal-Wallis test
statistic, p_value = kruskal(most_genre_EU, most_genre_NA, most_genre_JP, most_genre_Other)

print("Kruskal-Wallis Test:")
print(f"Test Statistic: {statistic}")
print(f"P-value: {p_value}")
```

After, knowing that all four of them are significantly different, we want to observe the pairwise comparisons using Dunn's Test with Bonferroni Correction for further analysis. It was found that all pairwise comparisons showed significant difference as well as the p-value is lower than 0.05, thus we fail to reject H0.

```python
# Retrieving the sales data from each region for action games
sales_data = [action_data['NA_Sales'], action_data['EU_Sales'], action_data['JP_Sales'], action_data['Other_Sales']]

# Performing the Kruskal-Wallis test
result = sp.posthoc_dunn(sales_data, p_adjust='bonferroni')
print("Dunn's Test with Bonferroni Correction:")
print(result)
```

After calculating the mean difference and its significance, the sales from each regions are presented in the chart below. The region with the highest mean difference is NA, followed by EU, Other, and JP.

```python
# Retrieving the sales average from each region for action games
mean_sales_by_region = action_data[['NA_Sales', 'EU_Sales', 'Other_Sales', 'JP_Sales']].mean() 

# Creating the bar chart
ax = mean_sales_by_region.plot(kind='bar', color='deeppink', figsize=(10, 6))
plt.xlabel('Region')
plt.ylabel('Mean Sales')
plt.title('Action Sales by Region on Average')

plt.show()
```

After that, we want to evaluate the correlation between critic score and global sales, as well as user score and global sales. The correlation analysis was conducted using Kendall Tau method, as the 'Global_Sales' column is continous whereas the 'User_Score' and 'Critic_Score' columns are ordinal. Thus, Kendall Tau method is the most appropirate approach for this case. After retrieving the correlation coefficient, scatter graphs of each analysis was plotted and correlation line was also drawn to represent the correlation.

Below are the hypotheses of this test:

1. H0: Global sales and critic score are not significantly correlated
2. H1: There is a significant correlation between global sales and critic score

and 

1. H0: Global sales and user score are not significantly correlated
2. H1: There is a significant correlation between global sales and user score

From the correlation analysis, it was found that critic score and global sales have a weak positive correlation with a value of 0.29. However, the p-value of is less than 0.05, which indicates that the correlation is statistically significant eventhough the strength is relatively weak and not particularly impactful. Therefore, the first H0 is rejected and we can conclude there is a weak correlation between critic score and global sales. 

Similar results were also found when analyzing the correlation between user score and global sales. However, the correlation coefficient is even lower, which only falls around 0.13. Despite having an even lower correlation, the correlation is still proven to be statistically significant (p <0.05). Therefore, the second H0 is rejected and we can conclude there is an extremely weak correlation between user score and global sales. 


From this analysis, we can conclude that having higher critic score will increase the likelihood of our video game generating higher sales. Thus, cooperating with critics to review our games will be a good strategy to maximize our sales. However, it is important to note that correlation does not mean causation, especially when the correlation is this low. 

```python
#Finding the Kendall correlation between global sales and critic score
correlation_critic, p_value_critic = stats.kendalltau(action_data["Critic_Score"], action_data["Global_Sales"][action_data["Critic_Score"].notnull()])
print(f"Critic Score vs Global Sales: Correlation = {correlation_critic}, P-value = {p_value_critic}")

print('================================================================================================')

#Finding the Kendall correlation between global sales and user score
correlation_user, p_value_user = stats.kendalltau(action_data["User_Score"], action_data["Global_Sales"][action_data["User_Score"].notnull()])
print(f"User Score vs Global Sales: Correlation = {correlation_user}, P-value = {p_value_user}")
```

After obtaining the correlations and p-values, scatter plot and regression line were created for better visualization

```python
# Creating Scatter Plot
plt.scatter(action_data["Critic_Score"], action_data["Global_Sales"][action_data["Critic_Score"].notnull()], color='deeppink') 

coefficients = np.polyfit(action_data["Critic_Score"], action_data["Global_Sales"][action_data["Critic_Score"].notnull()], 1)
slope = coefficients[0]
intercept = coefficients[1]

# Drawing the correlation line
regression_line = slope * np.array(action_data["Critic_Score"]) + intercept
plt.plot(action_data["Critic_Score"], regression_line, color='black', label=f'Correlation line: {correlation_critic:.2f}, p: {p_value_critic:.2f}')
plt.title('Correlation Global Sales vs Critic Score')
plt.xlabel('Critic Score')
plt.ylabel('Global Sales')
plt.legend()
plt.show()
```

```python
# Creating Scatter Plot
plt.scatter(action_data["User_Score"], action_data["Global_Sales"][action_data["User_Score"].notnull()], color='deeppink')

coefficients = np.polyfit(action_data["User_Score"], action_data["Global_Sales"][action_data["User_Score"].notnull()], 1)
slope = coefficients[0]
intercept = coefficients[1]

# Drawing the correlation line
regression_line = slope * np.array(action_data["User_Score"]) + intercept
plt.plot(action_data["User_Score"], regression_line, color='black', label=f'Correlation line: {correlation_user:.2f} , p: {p_value_user:.2f}')
plt.title('Correlation Global Sales vs User Score')
plt.xlabel('User Score')
plt.ylabel('Global Sales')
plt.legend()
plt.show()
```
